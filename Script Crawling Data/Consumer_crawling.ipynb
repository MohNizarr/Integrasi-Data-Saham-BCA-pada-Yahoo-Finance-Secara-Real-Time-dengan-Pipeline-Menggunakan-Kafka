```python
import csv
from kafka import KafkaConsumer
import json

# Inisialisasi Kafka Consumer
consumer = KafkaConsumer('stock-data', bootstrap_servers='localhost:9092')

# Membuka file CSV untuk penulisan data
csv_file = open('stock_data.csv', 'w', newline='')
csv_writer = csv.writer(csv_file)

# Menulis header kolom ke file CSV
csv_writer.writerow(['Date', 'Open', 'High', 'Low', 'Close', 'Volume'])

# Membaca dan menyimpan data saham dari Kafka ke dalam file CSV
for message in consumer:
    stock_data = json.loads(message.value.decode('utf-8'))
    csv_writer.writerow([stock_data['date'], stock_data['open'], stock_data['high'], stock_data['low'], stock_data['close'], stock_data['volume']])  

    # Menampilkan data yang disimpan ke CSV
    print("Data Tersimpan di CSV:")
    print("Date:", stock_data['date'])
    print("Open:", stock_data['open'])
    print("High:", stock_data['high'])
    print("Low:", stock_data['low'])
    print("Close:", stock_data['close'])
    print("Volume:", stock_data['volume'])
    print()

# Menutup file CSV dan koneksi Kafka setelah selesai
csv_file.close()
consumer.close()
```
